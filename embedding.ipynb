```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?',
    'i am good developer'
]
voc_size=10000
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_lenght=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_lenght)
print(embedded_docs)
```

```python
## feature representation
dim=10
model=Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_lenght))
model.build(input_shape=(None, sent_lenght))  # ensure weights exist so summary shows params
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
